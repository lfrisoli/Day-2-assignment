```python
sales = float(input("Enter sales ($): "))
tax_rate = float(input("Enter tax rate (%): "))

cogs = sales * 0.40
gross_profit = sales - cogs
taxes = gross_profit * (tax_rate / 100)
net_income = gross_profit - taxes

print("\nIncome Statement")
print(f"Sales: $ {sales:,.2f}")
print(f"COGS: $ {cogs:,.2f}")
print(f"Gross Profit: $ {gross_profit:,.2f}")
print(f"Taxes: $ {taxes:,.2f}")
print(f"Net Income: $ {net_income:,.2f}")
```

```python
def roi(cost, revenue):
    if not isinstance(cost,(int,float)) or not isinstance(revenue,(int,float)):
        return "Invalid input"
    if cost <= 0:
        return "Cost must be > 0"
    return (revenue - cost)/cost

print("ROI:", roi(100,120))
```

```python
def loan_payment(P, annual_rate, years):
    if P<=0 or annual_rate<=0 or years<=0:
        return "Invalid"
    r = annual_rate/12
    n = years*12
    return (r*P)/(1-(1+r)**(-n))

print("Payment:", loan_payment(10000,0.05,5))
```

```python
def getValues(n):
    lst=[]
    for i in range(n):
        while True:
            try:
                x=float(input("Value: "))
                if x<0: raise ValueError
                lst.append(x)
                break
            except:
                print("Invalid")
    return lst

def calcAverage(lst):
    return sum(lst[3:])/len(lst[3:])

vals = getValues(5)
print("Avg (excluding first 3):", calcAverage(vals))
```

```python
class Greeting:
    def __init__(self, first, last):
        self.first = first
        self.last = last

    def display(self, formal):
        if formal:
            print(f"{self.last}, {self.first}")
        else:
            print(self.first)

    def __str__(self):
        return f"{self.first} {self.last}"

g = Greeting("Leo","Frisoli")
g.display(True)
print(g)
```

```python
class Student:
    def __init__(self,name,num):
        self.name=name
        self.number=num
        self.courses=[]

    def get_name(self): return self.name
    def get_courses(self): return self.courses
    def set_name(self,n): self.name=n
    def __len__(self): return len(self.courses)

    def enroll(self,course):
        if course not in self.courses:
            self.courses.append(course)

s=Student("Leo",1)
s.enroll("Finance")
print(len(s))
```

```python
class Point:
    def __init__(self,x,y):
        self.x=x; self.y=y

    def distance(self,other):
        return math.sqrt((self.x-other.x)**2+(self.y-other.y)**2)

class Rectangle:
    def __init__(self,p1,p2,p3,p4):
        self.p1=p1; self.p2=p2; self.p3=p3; self.p4=p4

    def width(self):
        return self.p1.distance(self.p2)

    def height(self):
        return self.p2.distance(self.p3)

    def area(self):
        return self.width()*self.height()
```

```python
class Employee:
    def __init__(self,name,wage,hours):
        self.name=name
        self.__wage=wage
        self.hours=hours

    def get_wage(self): return self.__wage
    def set_wage(self,w):
        if w>=0: self.__wage=w

    def calculate_pay(self):
        return self.__wage*self.hours

    def display_info(self):
        print(self.name,self.calculate_pay())

class Manager(Employee):
    def __init__(self,name,wage,hours,bonus):
        super().__init__(name,wage,hours)
        self.bonus=bonus

    def calculate_pay(self):
        return super().calculate_pay()+self.bonus


e=Employee("John",20,40)
m=Manager("Sarah",30,40,500)

e.display_info()
m.display_info()
```
