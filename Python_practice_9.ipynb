```python
sales = float(input("Enter sales ($): "))
tax_rate_percent = float(input("Enter tax rate (%): "))

cogs = sales * 0.40
gross_profit = sales - cogs
taxes = gross_profit * (tax_rate_percent / 100)
net_income = gross_profit - taxes

print("\nIncome Statement")
print(f"Sales: $ {sales:,.2f}")
print(f"Cost of Goods Sold: $ {cogs:,.2f}")
print(f"Gross Profit: $ {gross_profit:,.2f}")
print(f"Taxes: $ {taxes:,.2f}")
print(f"Net Income: $ {net_income:,.2f}")


# -----------------------------
# 2. EOQ (Inventory Management)
# -----------------------------
import math

demand = float(input("\nEnter annual demand: "))
order_cost = float(input("Enter ordering cost ($): "))
holding_cost = float(input("Enter holding cost per unit ($): "))

eoq = math.sqrt((2 * demand * order_cost) / holding_cost)
print(f"Economic Order Quantity: {eoq:.2f}")


# -----------------------------
# 3. Calculator
# -----------------------------
num1 = float(input("\nEnter first number: "))
num2 = float(input("Enter second number: "))
operator = input("Enter operator (+, -, *, /): ")

if operator == "+":
    print("Answer:", num1 + num2)
elif operator == "-":
    print("Answer:", num1 - num2)
elif operator == "*":
    print("Answer:", num1 * num2)
elif operator == "/":
    if num2 != 0:
        print("Answer:", num1 / num2)
    else:
        print("Error: Cannot divide by zero.")
else:
    print("Invalid operator")


# -----------------------------
# 4. Parking Garage
# -----------------------------
print("\nHours  Fee")
for hours in range(1, 9):
    fee = 5 + (2.5 * hours)
    print(hours, fee)


# BONUS VERSION
print("\nBonus (Min $10, Max $20)")
for hours in range(1, 9):
    fee = 5 + (2.5 * hours)
    if fee < 10:
        fee = 10
    elif fee > 20:
        fee = 20
    print(hours, fee)


# -----------------------------
# 5. Movie Theater Profit Table
# -----------------------------
print("\nAd$   Profit")

for ad in range(0, 201, 25):
    additional_attendees = 2 * (ad ** 0.5)
    total_attendees = 20 + additional_attendees
    revenue = total_attendees * 10
    profit = revenue - ad - 200
    print(f"{ad:.1f}  {profit:.1f}")


# -----------------------------
# 6. Rock-Paper-Scissors
# -----------------------------
import random

while True:
    play = input("\nDo you want to play Rock-Paper-Scissors? (yes/no): ").lower()

    if play != "yes":
        print("Game ended.")
        break

    user = input("Choose rock, paper, or scissors: ").lower()
    computer = random.choice(["rock", "paper", "scissors"])

    print("Computer chose:", computer)

    if user == computer:
        print("It's a tie!")
    elif (user == "rock" and computer == "scissors") or \
         (user == "paper" and computer == "rock") or \
         (user == "scissors" and computer == "paper"):
        print("You win!")
    else:
        print("You lose!")
```
