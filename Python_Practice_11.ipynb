```python
import math
import random
```

```python
sales = float(input("Enter sales ($): "))
tax_rate = float(input("Enter tax rate (%): "))

cogs = sales * 0.40
gross_profit = sales - cogs
taxes = gross_profit * (tax_rate / 100)
net_income = gross_profit - taxes

print("\nIncome Statement")
print(f"Sales: $ {sales:,.2f}")
print(f"Cost of Goods Sold: $ {cogs:,.2f}")
print(f"Gross Profit: $ {gross_profit:,.2f}")
print(f"Taxes: $ {taxes:,.2f}")
print(f"Net Income: $ {net_income:,.2f}")
```

```python
demand = float(input("\nEnter annual demand: "))
order_cost = float(input("Enter ordering cost ($): "))
holding_cost = float(input("Enter holding cost per unit ($): "))

eoq = math.sqrt((2 * demand * order_cost) / holding_cost)
print(f"Economic Order Quantity: {eoq:.2f}")
```

```python
num1 = float(input("\nEnter first number: "))
num2 = float(input("Enter second number: "))
operator = input("Enter operator (+, -, *, /): ")

if operator == "+":
    print("Answer:", num1 + num2)
elif operator == "-":
    print("Answer:", num1 - num2)
elif operator == "*":
    print("Answer:", num1 * num2)
elif operator == "/":
    if num2 != 0:
        print("Answer:", num1 / num2)
    else:
        print("Error: Cannot divide by zero")
else:
    print("Invalid operator")
```

```python
print("\nParking Garage")
for hours in range(1, 9):
    fee = 5 + (2.5 * hours)
    print(hours, fee)

print("\nParking Garage Bonus")
for hours in range(1, 9):
    fee = 5 + (2.5 * hours)
    if fee < 10:
        fee = 10
    elif fee > 20:
        fee = 20
    print(hours, fee)
```

```python
print("\nAd$ Profit")
for ad in range(0, 201, 25):
    attendees = 20 + 2 * (ad ** 0.5)
    revenue = attendees * 10
    profit = revenue - ad - 200
    print(f"{ad:.1f} {profit:.1f}")
```

```python
while True:
    play = input("\nPlay Rock-Paper-Scissors? (yes/no): ").lower()
    if play != "yes":
        break

    user = input("Choose rock, paper, or scissors: ").lower()
    computer = random.choice(["rock", "paper", "scissors"])

    print("Computer:", computer)

    if user == computer:
        print("Tie")
    elif (user == "rock" and computer == "scissors") or \
         (user == "paper" and computer == "rock") or \
         (user == "scissors" and computer == "paper"):
        print("You win")
    else:
        print("You lose")
```

```python
numbers = []

while True:
    num = float(input("\nEnter number (0 to stop): "))
    if num == 0:
        break
    numbers.append(num)

numbers.sort(reverse=True)
print("Descending:", numbers)
```

```python
weights = [10.0, 9.9, 10.1, 10.05]

avg = sum(weights) / len(weights)
largest = max(weights)

print("\nAverage =", avg)
print("Largest =", largest)
```

```python
counts = {}
N = 5

for i in range(N):
    name = input("\nEnter name: ")
    if name in counts:
        counts[name] += 1
    else:
        counts[name] = 1

print("Dictionary:", counts)
```

```python
products = {"P-01": 9.95, "P-02": 10.50}

pid = input("\nEnter product ID: ")

if pid in products:
    print(pid, products[pid])
else:
    print(pid, "not found")
```

```python
def reverse(s):
    if len(s) == 0:
        return s
    return reverse(s[1:]) + s[0]


text = input("\nEnter a string to reverse: ")
print("Reversed:", reverse(text))
```
