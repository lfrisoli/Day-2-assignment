```python
sales = float(input("Enter sales ($): "))
tax_rate = float(input("Enter tax rate (%): "))

cogs = sales * 0.40
gross_profit = sales - cogs
taxes = gross_profit * (tax_rate / 100)
net_income = gross_profit - taxes

print("\nIncome Statement")
print(f"Sales: $ {sales:,.2f}")
print(f"Cost of Goods Sold: $ {cogs:,.2f}")
print(f"Gross Profit: $ {gross_profit:,.2f}")
print(f"Taxes: $ {taxes:,.2f}")
print(f"Net Income: $ {net_income:,.2f}")
```

```python
num1 = float(input("\nEnter first number: "))
num2 = float(input("Enter second number: "))
operator = input("Enter operator (+, -, *, /): ")

if operator == "+":
    print("Answer:", num1 + num2)
elif operator == "-":
    print("Answer:", num1 - num2)
elif operator == "*":
    print("Answer:", num1 * num2)
elif operator == "/":
    if num2 != 0:
        print("Answer:", num1 / num2)
    else:
        print("Error: Cannot divide by zero")
else:
    print("Invalid operator")
```

```python
print("\nParking Garage")
print("Hours Fee")

for hours in range(1, 9):
    fee = 5 + (2.5 * hours)
    print(hours, fee)


# Parking Garage Bonus
print("\nParking Garage Bonus")
print("Hours Fee")

for hours in range(1, 9):
    fee = 5 + (2.5 * hours)

    if fee < 10:
        fee = 10
    elif fee > 20:
        fee = 20

    print(hours, fee)
```

```python
print("\nMovie Theater")
print("Ad$ Profit")

for ad in range(0, 201, 25):
    additional_attendees = 2 * (ad ** 0.5)
    total_attendees = 20 + additional_attendees
    revenue = total_attendees * 10
    profit = revenue - ad - 200

    print(f"{ad:.1f} {profit:.1f}")
```

```python
import random

play = input("\nDo you want to play Rock-Paper-Scissors? yes/no: ").lower()

while play == "yes":
    user = input("Choose rock, paper, or scissors: ").lower()
    computer = random.choice(["rock", "paper", "scissors"])

    print("Computer chose:", computer)

    if user == computer:
        print("Tie")
    elif user == "rock" and computer == "scissors":
        print("You win")
    elif user == "paper" and computer == "rock":
        print("You win")
    elif user == "scissors" and computer == "paper":
        print("You win")
    elif user == "rock" or user == "paper" or user == "scissors":
        print("You lose")
    else:
        print("Invalid choice")

    play = input("Do you want to play again? yes/no: ").lower()
```

```python
numbers = []

num = float(input("\nEnter a number, or 0 to stop: "))

while num != 0:
    numbers.append(num)
    num = float(input("Enter a number, or 0 to stop: "))

numbers.sort(reverse=True)
print("Numbers in descending order:", numbers)
```

```python
counts = {}

N = 5

for i in range(N):
    name = input("\nEnter a name: ")

    if name in counts:
        counts[name] = counts[name] + 1
    else:
        counts[name] = 1

print("Final dictionary:", counts)
```
