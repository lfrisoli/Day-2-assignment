```python
sales = float(input("Enter sales ($): "))
tax_rate_percent = float(input("Enter tax rate (%): "))

cogs = sales * 0.40
gross_profit = sales - cogs
taxes = gross_profit * (tax_rate_percent / 100)
net_income = gross_profit - taxes

print("\nIncome Statement")
print(f"Sales: $ {sales:,.2f}")
print(f"Cost of Goods Sold: $ {cogs:,.2f}")
print(f"Gross Profit: $ {gross_profit:,.2f}")
print(f"Taxes: $ {taxes:,.2f}")
print(f"Net Income: $ {net_income:,.2f}")


# -----------------------------
# 2. Inventory Management EOQ
# -----------------------------

import math

demand = float(input("\nEnter annual demand: "))
order_cost = float(input("Enter ordering cost ($): "))
holding_cost = float(input("Enter annual holding cost per unit ($): "))

eoq = math.sqrt((2 * demand * order_cost) / holding_cost)

print(f"Economic Order Quantity: {eoq:.2f} units")


# -----------------------------
# 3. Numerical Calculator
# -----------------------------

num1 = float(input("\nEnter first number: "))
num2 = float(input("Enter second number: "))
operator = input("Enter operator (+, -, *, /): ")

if operator == "+":
    answer = num1 + num2
    print("Answer:", answer)
elif operator == "-":
    answer = num1 - num2
    print("Answer:", answer)
elif operator == "*":
    answer = num1 * num2
    print("Answer:", answer)
elif operator == "/":
    if num2 != 0:
        answer = num1 / num2
        print("Answer:", answer)
    else:
        print("Error: Cannot divide by zero.")
else:
    print("Invalid operator.")
```
